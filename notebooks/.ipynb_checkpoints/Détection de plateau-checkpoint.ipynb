```python
import numpy as np
import cv2 as cv2
from matplotlib import pyplot as plt
```

```python
def draw_lines(img, lines, r, g, b, e = 1):
    
    if lines is None:
        return img
    
    for r_theta in lines:
        arr = np.array(r_theta[0], dtype=np.float64)
        r, theta = arr
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*r
        y0 = b*r
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(img, (x1, y1), (x2, y2), (r, g, b), e)
        
    return img
```

```python
def showChess(img):

    # pre-traitement de l'image
    
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    img_median = cv2.medianBlur(gray, 7)
    img_median = cv2.dilate(img_median, None, iterations = 3)
    img_median = cv2.erode(img_median, None, iterations = 3)

    threshold_img = cv2.adaptiveThreshold(img_median,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    threshold_img = cv2.bitwise_not(threshold_img)
    threshold_img = cv2.medianBlur(threshold_img, 5)

    threshold_img = cv2.dilate(threshold_img, None, iterations = 6)
    threshold_img = cv2.erode(threshold_img, None, iterations = 3)

    # Detection de contour
    
    canny_img = cv2.Canny(threshold_img,150,255)
    contours, hierarchy = cv2.findContours(canny_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    # Recherche du meilleur contour
    
    max_find = 0
    max_cnt = None

    for cnt in contours:        
        rect = cv2.minAreaRect(cnt)
        width = rect[1][0]
        height = rect[1][1]

        area = width * height

        if area > max_find:
            max_find = area
            max_cnt = cnt

    if max_cnt is None:
        return img
    
    (x,y,w,h) = cv2.boundingRect(max_cnt)
    cv2.rectangle(img, (x,y), (x+w,y+h), (255, 255, 0), 1)
    cv2.drawContours(img, [max_cnt], -1, (255,0,0), 1)
    
    
    # Recherche de ligne
    
    black = img.copy()
    black[True] = 0

    contour_lines = cv2.drawContours(black, [max_cnt], -1, (255, 255, 255),1)
    contour_lines = cv2.cvtColor(contour_lines,cv2.COLOR_BGR2GRAY)

    lines_second = cv2.HoughLines(contour_lines, 1, np.pi/180, 50)
    lines = cv2.HoughLines(contour_lines, 1, np.pi/180, 80)
    
    # draw_lines(img, lines_second, 0, 255, 0)
    draw_lines(img, lines, 0, 0, 255)
   
    return img
```

```python
def mask_chess(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    img_median = cv2.medianBlur(gray, 7)
    img_median = cv2.dilate(img_median, None, iterations = 10)
    
    return img_median

    im_th = cv2.adaptiveThreshold(img_median,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    im_th = cv2.bitwise_not(im_th)
    im_th = cv2.medianBlur(im_th, 5)

    im_th = cv2.dilate(im_th, None, iterations = 6)
    im_th = cv2.erode(im_th, None, iterations = 3)

    canny_img = cv2.Canny(im_th,150,255)
    contours, hierarchy = cv2.findContours(canny_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    max_find = 0
    
    for cnt in contours:        
        rect = cv2.minAreaRect(cnt)       #I have used min Area rect for better result
        width = rect[1][0]
        height = rect[1][1]

        area = width * height

        if area > max_find:
            max_find = area
            max_cnt = cnt
    
    try: max_cnt
    except NameError: return img
    
    
    (x,y,w,h) = cv2.boundingRect(max_cnt)
    
    max_x = img.shape[1] - 1
    max_y = img.shape[0] - 1
    
    fill_x = 0
    fill_y = 0
    
    if max_x - (x + w) > x:
        fill_x = max_x - 1
    if max_y - (y + h) > y:
        fill_y = max_y - 1
    
    
    
    # Lines
    black = img.copy()
    black[True] = 0

    contour_lines = cv2.drawContours(black, [max_cnt], -1, (255, 255, 255),1)
    contour_lines = cv2.cvtColor(contour_lines,cv2.COLOR_BGR2GRAY)

    im_th = contour_lines
    im_floodfill = contour_lines.copy()

    h, w = im_th.shape[:2]
    mask = np.zeros((h+2, w+2), np.uint8)

    cv2.floodFill(im_floodfill, mask, (fill_x, fill_y), 255);

    im_floodfill_inv = cv2.bitwise_not(im_floodfill)
    im_floodfill_inv = cv2.erode(im_floodfill_inv, None, iterations = 15)
    
    return im_floodfill_inv
```

```python
img = cv2.imread('Images/plateau avec piece 1.png')

print(img.shape)

img_2 = showChess(img)
plt.figure(figsize=(4,4))
plt.imshow(img_2[:,:,::-1],cmap = 'gray')
```
